```python
#예측값으로 올바른 모델 찾기
```

```python
w = 1.0
b = 1.0
```

```python
y_hat = x[0] * w + b
print(y_hat)
```

```python
print(y[0])
```

```python
# w값을 조절해 예측값 바꾸기
w_inc = w + 0.1
y_hat_inc = x[0] * w_inc + b
print(y_hat_inc)
```

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
```

```python
# 변화율로 절편 업데이트하기
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)

b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

```python
#변화율로 절편 업데이트트
```

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)
```

```python
b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

```python
b_new = b + 1
print(b_new)
```

```python
#오차와 변화율을 곱하여 가중치 업데이트하기
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
```

```python
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
print(w_new, b_new)
```

```python
for x_i, y_i in zip(x, y):
  y_hat = x_i * w + b
  err = y_i - y_hat
  w_rate = x_i
  w = w+w_rate * err
  b = b + 1*err
print(w, b)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#에포크 진행
for i in range(1, 100):
  for x_i, y_i in zip(x, y):
    y_hat = x_i * w + b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate * err
    b = b + 1 * err

print(w, b)
```
