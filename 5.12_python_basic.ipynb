```python
import pandas as pd
marathon_2015 = pd.read_csv('./marathon_results_2015.csv')
marathon_2016 = pd.read_csv('./marathon_results_2016.csv')
marathon_2017 = pd.read_csv('./marathon_results_2017.csv')
```

```python
marathon_2017.head()
```

```python
marathon_2015['Year'] = '2015'
marathon_2016['Year'] = '2016'
marathon_2017['Year'] = '2017'
```

```python
marathon_2015_2017.info()
```

```python
marathon_2015_2017 = marathon_2015_2017.drop(['Unnamed: 0','Bib','Citizen','Unnamed: 9', 'Proj Time','Unnamed: 8'], axis=1)
```

```python
import pandas as pd
marathon_2015_2017 = pd.read_csv('./marathon_2015_2017.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marathon_2015_2017.head()
```

```python
USA_runner = marathon_2015_2017[marathon_2015_2017.Country == 'USA']
USA_runner.info()
```

```python
plt.figure(figsize=(300,20))
```

```python
runner_state = sns.countplot(x='State', data=USA_runner)
runner_state.set_title('Number of Runner by State - USA', fontsize=18)
runner_state.set_xlabel('State', fontdict={'size':16})
runner_state.set_ylabel('Number of Runner', fontdict={'size':16})
plt.show()
```

```python
plt.figure(figsize=(30,20))
runner_state = sns.countplot(x='State', data=USA_runner, hue='M/F', palette={'F':'r','M':'b'})
runner_state.set_title('Number of Runner by State, Gender - USA', fontsize=18)
runner_state.set_xlabel('State', fontdict={'size':16})
runner_state.set_ylabel('Number of Runner', fontdict={'size':16})
plt.show()
```

```python
plt.figure(figsize=(30,20))
runner_state = sns.countplot(x='State', data=USA_runner, hue='Year')
runner_state.set_title('Number of Runner by State, year - USA', fontsize=18)
runner_state.set_xlabel('State', fontdict={'size':16})
runner_state.set_ylabel('Number of Runner', fontdict={'size':16})
plt.show()
```

```python
import pandas as pd
marathon_2015_2017 = pd.read_csv('./marathon_2015_2017.csv')
import matplotlib.pyplot as plt
```

```python
runner_1860 = marathon_2015_2017[marathon_2015_2017.Age.isin(range(18,60))]
runner_1860_counting = runner_1860['Age'].value_counts()
```

```python
runner_1860.info()
```

```python
runner_1860_counting
```

```python
x = runner_1860_counting.index
x = [str(i) for i in x]
y = runner_1860_counting.values

ratio = y/y.sum()
ratio_sum = ratio.cumsum()
```

```python
fig, barChart = plt.subplots(figsize=(20,10))
barChart.bar(x,y)
lineChart = barChart.twinx()
lineChart.plot(x, ratio_sum, '-ro', alpha=0.5)
ranges = lineChart.get_yticks()
lineChart.set_yticklabels(['{0:01%}.format(x) for x in ranges'])
```

```python
import pandas as pd
marathon_2015_2017 = pd.read_csv('./marathon_2015_2017.csv')
import matplotlib.pyplot as plt
record = pd.DataFrame(marathon_2015_2017, columns=['5K','10K','20K','Half','25K','30K','35K','40K','Official Time']).sort_values(by=['Official Time'])
```

```python
record.info()
```

```python
record.head()
```

```python
record.insert(0,'Rank',range(1,1+len(record)))
```

```python
record.head()
```

```python
top100 = record[0:100]
```

```python
xData = top100.Rank
```

```python
xData
```

```python
yData_final = top100['Official Time']
yData_10K = top100['10K']
yData_20K = top100['20K']
yData_30K = top100['30K']

yData_final
```

```python
plt.figure(figsize=(20,10))
plt.plot(xData, yData_10K)
plt.plot(xData, yData_20K)
plt.plot(xData, yData_30K)
plt.plot(xData, yData_final)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(xData, yData_10K, 'r--')

plt.subplot(2,2,2)
plt.plot(xData, yData_20K, 'g:')

plt.subplot(2,2,3)
plt.plot(xData, yData_30K, 'y--')

plt.subplot(2,2,4)
plt.plot(xData, yData_final, 'b-.')

plt.show()
```

```python
import pandas as pd
marathon_2015_2017 = pd.read_csv('./marathon_2015_2017.csv')
import matplotlib.pyplot as plt
```

```python
MALE_runner = marathon_2015_2017[marathon_2015_2017['M/F'] == 'M']
FEMALE_runner = marathon_2015_2017[marathon_2015_2017['M/F'] == 'F']
```

```python
FEMALE_runner.head()
```

```python
plt.figure(figsize=(20,20))
x_male = MALE_runner.Age
y_male = MALE_runner['Official Time']
x_female = FEMALE_runner.Age
y_female = FEMALE_runner['Official Time']
```

```python
x_male
```

```python
y_male
```

```python
plt.figure(figsize=(20,20))
plt.plot(x_male, y_male, '.b', alpha=0.5)
plt.plot(x_female, y_female, '.r', alpha=0.5)

plt.xlabel('Age', fontsize=16)
plt.ylabel('Official Time (Second)', fontsize=16)
plt.title('Distribution by Running Time and Age', fontsize=20)

plt.show()
```
